```python
import yfinance as yf
import ta
def get_price_data(stock_name,start='2018-01-01',end='2022-12-31'):
  stock = yf.Ticker(stock_name)
  history = stock.history(start= start,end = end)
  if history.empty:
    print("Check if entered stock name ",stock_name," is valid")
  else:
    indicators = ta.add_all_ta_features(history, open="Open",close="Close",high="High",low="Low",volume="Volume")
    return indicators
```

```python
price_data = get_price_data("^DJI")
price_data.head()
```
